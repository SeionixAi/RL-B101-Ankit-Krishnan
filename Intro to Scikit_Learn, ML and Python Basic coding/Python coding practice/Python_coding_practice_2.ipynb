1. Python program to check whetehr a string contains only digits

```python
def is_only_digits(text):
  for char in text:
    if not char.isdigit():
      return False
  return f"contains only digits"


text= input("Enter a string: ")
print(is_only_digits(text))
```

2. Python program to find longest word in a string

```python
def find_longest_word(text):
  words = text.split()
  longest_word=""
  for word in words:
    if len(word) > len(longest_word):
      longest_word=word
  return longest_word

text= input("Enter a sentence: ")
print(find_longest_word(text))
```

3. Python program to capitalize first and last chars of each word in a string

```python
def capitalize_ends(text):
  words = text.split()
  result = []

  for word in words:
    if len(word) > 1:
      new_word = word[0].upper() + word[1:-1] + word[-1].upper()
      result.append(new_word)
    else:
      result.append(word.upper())

  return" ".join(result)

text=input("Enter a sentence: ")
print(capitalize_ends(text))
```

4. Draw a right angled triangle using stars

```python
def right_triangle(n):
  for i in range(1,n+1):
    print("*"*i)
```

```python
right_triangle(4)
```

5. draw an inverted right angled triangle pattern

```python
def inverted_triangle(n):
  for i in range(n,0,-1):
    print("*"*i)
```

```python
inverted_triangle(4)
```
