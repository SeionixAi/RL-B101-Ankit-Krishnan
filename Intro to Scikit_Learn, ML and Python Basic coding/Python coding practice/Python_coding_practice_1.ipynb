1. Python program for counting number of consonants in a given string.

```python
def count_consonants(text):
  vowels = "aeiou"
  count = 0
  for char in text.lower():
    if char.isalpha() and char not in vowels:
      count += 1
  return count

text = input("Enter a text:")
print("Number of consonants:", count_consonants(text))
```

2. Python program to remove all duplicate chars from a string

```python
def remove_duplicates(text):
  result=""
  for char in text.lower():
    if char not in result:
      result += char
  return result

text=input("Enter a text:")
print(remove_duplicates(text))
```

3. Python program to count frequency of each char in a string

```python
def count_char_frequency(text):
  frequency={}
  for char in text.lower():
    if char in frequency:
      frequency[char] += 1
    else:
      frequency[char] = 1
  return frequency

text=input("Enter a text:")
print(count_char_frequency(text))
```

4. Python program to find first non-repeating char in a string

```python
def first_non_repeating(text):
  lower_text=text.lower()
  char_count={}
  for char in lower_text:
    if char in char_count:
      char_count[char]+=1
    else:
      char_count[char]=1

  for char in lower_text:
    if char_count[char]==1:
      return char
  return char


text=input("Enter a text:")
print(first_non_repeating(text))

```
