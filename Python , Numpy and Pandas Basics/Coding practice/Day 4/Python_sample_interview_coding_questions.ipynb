1. Python program to count number of vowels in a given string

```python
def count_vowels(text):
  vowels="aeiou"
  count=0
  for char in text.lower():
    if char in vowels:
      count+=1
  return count

print(count_vowels("Artificial Intelligence"))
```

```python
given_text=input("Enter a text: ")

print(count_vowels(given_text))
```

2. Python Program to check if a given string is palindrome

```python
def is_palindrome(text):
  #convert text to lower
  lower_text=text.lower()

  #reverse lowercase text
  reversed_text=lower_text[::-1]

  #compare original and reversed
  if lower_text==reversed_text:
    return True
  else:
    return False

print(is_palindrome("Malayalam"))
```

```python
text_given=input("enter a text: ")
print(is_palindrome(text_given))
```

3. Python program to check if 2 strings are anagrams

```python
def is_anagram(txt1, txt2):
  #convert to lower
  text1=txt1.lower()
  text2=txt2.lower()

  #sort chars of both text
  sorted_text1=sorted(text1)
  sorted_text2=sorted(text2)

  #compare sorted chars list
  if sorted_text1==sorted_text2:
    return True
  else:
    return False


character1=input("Enter a text: ")
character2=input("Enter the next text: ")
print("whether they are anagrams: ", is_anagram(character1,character2))
```
