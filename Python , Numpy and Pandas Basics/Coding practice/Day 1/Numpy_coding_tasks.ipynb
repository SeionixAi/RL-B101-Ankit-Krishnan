1. Creating arrays

```python
import numpy as np

arr=np.array([1,2,3,4,5])
print(arr)
```

2. Array operations

```python
a=np.array([10,20,30])
b=np.array([1,2,3])

print(a+b)
print(a*b)
print(a-b)
print(a/b)
```

3. Zeros and Ones

```python
print(np.zeros((2,3)))
print(np.ones((3,3)))
```

4. Mean, Max, Min

```python
data=np.array([10,20,30,40])

print("Mean: ", np.mean(data))
print("Max: ", np.max(data))
print("Min: ", np.min(data))
```
