1. Create DataFrame

```python
import pandas as pd

data={
    "Name":["Alice","Bob","Charlie"],
    "Age":[22,24,23],
    "Marks":[85,90,88]
}
df=pd.DataFrame(data)
print(df)

df.to_csv('data.csv', index=False)
print('DataFrame saved to data.csv')
```

2. Read CSV

```python
df = pd.read_csv("data.csv")
print(df.head())
```

3. Basic operations

```python
print(df["Age"].mean())
print(df["Marks"].max())
```

4. Filtering

```python
print(df[df["Marks"]>85])
```
