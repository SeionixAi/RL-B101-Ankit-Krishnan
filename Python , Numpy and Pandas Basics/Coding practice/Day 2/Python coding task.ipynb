1. Dynamic Typing

```python
y=13
print(type(y))

y="Hello World"
print(type(y))
```

2. Mutable vs Immutable data types

```python
l=[2,6,3,7]
l[2]=9
print(l)

t=(5,7,9,11)
#t[1]=4
print(t)
```

3. Local and Global Variables

```python
age=44

def man():
    name="Manesh"
    print(age,name)

man()

print(age)
#print(name)
```

4. Return Statement in a Function

```python
def multiply(a,b):
    return a*b

result=multiply(4,8)
print(result)
```

5. List Comprehension

```python
squares=[x*x for x in range(1,11)]
print(squares)
```

6. Break, continue, and pass

```python
for i in range(1,21):
    if i==6:
        continue
    if i==9:
        pass
    if i==11:
        break
    print(i)
```

7. Recursion

```python
def factorial(n):
    if n==0:
        return 1
    return n*factorial(n-1)

print(factorial(7))
```

8. Lambda Function

```python
square=lambda x:x*x
print(square(5))
```

9. is vs ==

```python
a=[2,3,4]
b=[2,3,4]

print(a==b)
print(a is b)
```

10. Docstrings

```python
def greet():
    """This function prints a greeting message"""
    print("Hello Python")

greet()
print(greet._doc_)
```
